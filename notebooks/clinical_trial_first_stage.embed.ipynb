# First stage report

```
first_stage_data_raw <- read_sheet("https://docs.google.com/spreadsheets/d/12EZz5S6ummuC9x5kwpWkjX7-l5bErUbtRZeJufXLK_o/edit?usp=sharing", sheet = "stage1")
```

```
#| label: tbl-demo-stage1
#| echo: false
#| tbl-cap: "Demographics of the first stage"
first_stage_data_raw |>
  select(trt, outcome, bmi, diabetis) |>
  tbl_summary() 
```

```
#| label: fig-bmi
#| echo: false
ggstatsplot::ggbetweenstats(
  data = first_stage_data_raw,
  x = trt,
  y = bmi,
  pairwise.display = "p-value",
  p.adjust.method = "bonferroni",
  var.equal = TRUE,
  title = "BMI by treatment group"
)
```
